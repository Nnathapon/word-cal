```python
def s3(s):
    return [eval("0x" + x)/255 for x in (s[:2], s[2:4], s[4:])]
rgb = []
for i in clrlist:
    if "," in i:
        q = i.split(",")
        print(q)
        rgb += [s3(q[0])]
        rgb += [s3(q[1])]
    else:
        rgb += [s3(i)]
```
