```python
f = open("thai-wordlist.txt", "r")
words = f.readlines()
```

```python
prefix={}
count=0
for i in words:
    for j in words:
        if j.startswith(i.strip()) and i!=j:
            if i in prefix:
                prefix[i] += 1
            else:
                prefix[i] = 1
    count+=1
    if count%100==0:
        print(i)
```

```python
def sorter(l,fx):
    return dict(sorted(l.items(),key=fx))
```

```python
fix=sorter(prefix,lambda x:-(x[1]))
```

```python
fix=sorter(sorter(prefix,lambda x:-(x[1])),lambda x:-len(x[0]))
```

```python
print(fix)
```

```python
s=[]
for i in fix:
    s+=[i.strip()+' '+str(fix[i])+'\n']
print(s)
with open('prefix.txt','w') as f:
    f.writelines(s)
```
