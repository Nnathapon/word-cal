```python
from pythainlp.corpus import get_corpus
s=sorted(get_corpus('tnc_freq.txt'),key=lambda x:int(x.split('\t')[1]))[::-1]
```

```python
ch={}
co=0
for i in s:
    word,num=i.split('\t')
    for j in word:
        if j in ch:
            ch[j]+=int(num)
        else:
            ch[j]=int(num)
    co+=1
    if co%100==0:print(i)
print(ch)
```

```python
def sorter(l,fx):
    return dict(sorted(l.items(),key=fx))
```

```python
x=sorter(ch,lambda x:-x[1])
```

```python
x
```

```python
tot=sum(x.values())
a=[]
for i in x:
    a+=["%s, %f\n" % (i,x[i]*100/tot)]
a
```

```python
with open('freq.csv','w') as f:
    f.writelines(a)
```
